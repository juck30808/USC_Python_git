# 字串

```python
# 可以用單引號
'Hello world'
```

```python
# 也可以用雙引號
"Hello world"
```

```python
"Clement's dog"
```

```python
# 當字串中有單引號時的用法
'Clement\'s dog'
```

```python
# 字串可以使用 + 運算子連接 (concatenate)，並用 * 重覆該字串的內容
3 * 'un' + 'ium'
```

```python
# 兩個以上相鄰的字串值（string literal，即被引號包圍的字串）會被自動連接起來
'Py' 'thon'
```

```python
# 當你想要分段一個非常長的字串時，兩相鄰字串值自動連接的特性十分有用
text = ('Put several strings within parentheses '
    'to have them joined together.')
text
```

```python
# 但是變數跟字串不能直接連接
prefix = 'Py'
prefix 'thon'
```

```python
# 需要使用 + 這個運算子來連接
prefix = 'Py'
prefix + 'thon'
```

```python
# 字串可以被「索引 indexed」(下標，即 subscripted)，第一個字元的索引值為 0。
# 沒有獨立表示字元的型別；
# 一個字元就是一個大小為 1 的字串
word = 'Python'
word[0]
```

```python
word[5]
```

```python
# 超出範圍會有錯誤訊息
word[6]
```

```python
# 索引值可以是負的，此時改成從右開始計數
word[-1]
```

```python
word[-6]
```

```python
# 一樣超出範圍會有錯誤訊息
word[-7]
```

```python
# 除了索引外，字串亦支援「切片 slicing」。
# 索引用來拿到單獨的字元，而切片則可以讓你拿到子字串 (substring)
word[0:2]
```

```python
word[2:5]
```

```python
# 取出前兩個字元
word[:2]
```

```python
# 取出最後兩個字元
word[4:]
```

```python
# 取出最後兩個字元 (更容易的方式)
word[-2:]
```

```python
# Python 字串無法被改變 --- 它們是 immutable
word[0] = 'J'
```

```python
# 只能重新創造並組合成一個新字串
'J' + word[1:]
```

```python
# 取得字串長度
len(word)
```

# List 串列

```python
# List 像是一個陣列，儲存多個數值
squares = [1, 4, 9, 16, 25]
squares
```

```python
# 可以取出其中的數值
squares[0]
```

```python
# 也可以使用負數的索引
squares[-2]
```

```python
# 也可以取出其中一部分
squares[-3:]
```

```python
#所有的切片操作都返回一个包含所请求元素的新列表。 这意味着以下切片操作会返回列表的一个 浅拷贝
a = squares[:]
a[0] = 1111
print(a)
print(squares)
```

```python
# 可以連結兩個 list
squares + [36, 49, 64, 81, 100]
```

```python
# 跟字串不一樣的是 list 內的子元素是可以修改的
cubes = [1, 8, 27, 65, 125]
cubes[3] = 64
cubes
```

```python
# 新增一個新元素到list最後面
cubes.append(216)
```

```python
# 也可以對 slice 賦值，這能改變 list 的大小
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
letters[2:5] = ['C', 'D', 'E']
letters
```

```python
# 移除部分的元素
letters[2:5] = []
letters
```

```python
# 移除全部的元素
letters[:] = []
letters
```

```python
# 取得 list 的長度跟字串一樣方法
letters = ['a', 'b', 'c', 'd']
len(letters)
```

```python
# 可以嵌套多層 list （建立 list 包含其他 list）
a = ['a', 'b', 'c']
n = [1, 2, 3]
x = [a, n]
x
```

```python
x[0]
```

```python
# 等同於一個二維的矩陣
x[0][1]
```
