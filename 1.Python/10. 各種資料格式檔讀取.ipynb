```python
#HW1.讀取Iris.xlsx，分別計算 SepalLengthCm,SepalWidthCm,
#PetalLengthCm,PetalWidthCm 的平均值然後Print出來。

from openpyxl import load_workbook
wb = load_workbook('data/Iris.xlsx')
sheet = wb['Iris']

avgs = []
for column_id in range(2,6):
    sum = 0
    for row_id in range(2,sheet.max_row+1):
        sum += float(sheet.cell(row=row_id, column=column_id).value)
    avg = sum / (sheet.max_row - 1)
    avgs.append(avg)


print(avgs)
```

```python
#HW2.讀取Iris.xlsx，分別計算SepalLengthCm,SepalWidthCm,
#PetalLengthCm,PetalWidthCm的平均值、標準差跟四分位值然後將結果儲存在同一個檔案的新工作表中。

def countStatisticsValue(data):
    sum = 0
    for n in data:
        sum += n
    avg = sum / len(data)

    sum = 0
    for n in data:
        sum += (avg-n)*(avg-n)
    sd = (sum / len(data)) ** 0.5

    d2 = data[:]
    d2.sort()
    q1 = d2[int(len(d2)/4)]
    q2 = d2[int(len(d2)/2)]
    q3 = d2[int(len(d2)*3/4)]

    return (avg, sd, q1, q2, q3)


from openpyxl import load_workbook


wb = load_workbook('data/Iris.xlsx')
sheet = wb['Iris']

ans = []
for column_id in range(2,6):
    data = []
    for row_id in range(2,sheet.max_row+1):
        data.append(float(sheet.cell(row=row_id, column=column_id).value))
    result = countStatisticsValue(data)
    ans.append(result)
    
print(ans)
```

```python
#HW3.讀取hr_data.xlsx，統計department中總共有多少部門。

from openpyxl import load_workbook
wb = load_workbook('data/hr_data.xlsx')
sheet = wb.active

department_name = set()
for row_id in range(2,sheet.max_row+1):
    #同樣數值會過濾掉，把這個 Set 儲存不重複部門
    department_name.add(sheet.cell(row=row_id, column=10).value)

print(department_name)
```

```python
#HW4.讀取hr_data.xlsx，統計department中總共有多少部門，以及每個部門的資料人數。


from openpyxl import load_workbook
wb = load_workbook('data/hr_data.xlsx')
sheet = wb.active

department_name = {}
for row_id in range(2,sheet.max_row+1):
    dname = sheet.cell(row=row_id, column=10).value
    
    #檢查部門名稱(dname)是否有在department內，沒有的話定義1
    if dname in department_name:
        department_name[dname] += 1
    else:
        department_name[dname] = 1

print(department_name)
```

```python
#HW5.讀取hr_data.xlsx，average_montly_hours中列出最高前五名的加班時數值。
from openpyxl import load_workbook
wb = load_workbook('data/hr_data.xlsx')
sheet = wb.active

data = []
for row_id in range(2,sheet.max_row+1):
    #if blank
    try:
        data.append(int(sheet.cell(row=row_id, column=4).value))
    except:
        pass

data.sort(reverse=True)
top5 = data[0:5]

print(top5)
```

```python

#HW2-請計算 “國際主要國家貨幣每月匯率概況.json” 裡日圓的平均匯率與最低跟最高匯率


import json
f = open("data/國際主要國家貨幣每月匯率概況.json", "r", encoding='UTF-8')
jsonstr = f.read()
f.close()
data = json.loads(jsonstr)
print('日圓匯率')
sum = 0
max = 0
min = 9999999
for d in data:
    v = float(d['日圓'])
    if v > max:
        max = v
    if v < min:
        min = v
    sum += v
print('平均',sum / len(data))
print('最高', max)
print('最低', min)
```

```python
#請計算 “國際主要國家貨幣每月匯率概況.json” 裡所有幣別的平均匯率與最低跟最高匯率

import json

def 計算匯率(幣別, 資料):
    print(幣別+'匯率')
    sum = 0
    max = 0
    min = 9999999
    for d in 資料:
        v = float(d[幣別])
        if v > max:
            max = v
        if v < min:
            min = v
        sum += v
    print('平均',sum / len(資料))
    print('最高', max)
    print('最低', min)
    print('')

f = open("data/國際主要國家貨幣每月匯率概況.json", "r", encoding='UTF-8')
jsonstr = f.read()
f.close()
data = json.loads(jsonstr)

#方法一
計算匯率('新台幣', data)
計算匯率('人民幣', data)

#方法二
幣別群 = ["新台幣","人民幣","日圓","韓元","新加坡元","歐元","英鎊","澳幣"]
for 幣別 in 幣別群:
    計算匯率(幣別, data)

#方法三
# 幣別群 = data[0].keys()
# for 幣別 in 幣別群:
#     if 幣別 != '月別':
#         計算匯率(幣別, data)


```
