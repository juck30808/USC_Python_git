```python
import jieba
```

```python
# 準備斷詞資料
file = open("TEXT_01.txt", "r",encoding="utf-8")
text = file.read()
file.close()
print(text)
```

```python
# 開始斷詞
seg_list = jieba.cut(text, cut_all=False)
strs = list(seg_list)
print(strs)
```

```python
# 統計詞的出現次數
words = {}
for str in strs:
    if str in words:
        words[str] += 1
    else:
        words[str] = 1
        
for key, value in words.items():
    print(key, value)
```

```python
# 排序
sorted_keys = sorted(words, key=words.get, reverse=True)
for key in sorted_keys:
    print(key, words[key])
```
