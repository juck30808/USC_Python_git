# 資料正規化

```python
#示範用測試資料
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("data/heart.csv")
x = data.drop('target',axis=1)
y = data['target']
data.head(5)
```

## 為何要正規化

在很多演算法如KNN中會需要計算資料之間的距離。

先計算前兩筆資料的距離看看

```python
x1 = x.iloc[0]
x2 = x.iloc[1]
dist = np.sum((x1-x2)**2)*0.5
print('資料距離=',dist)
```

計算出來的距離值很大，但是每個欄位貢獻度都相容

```python
#譬如年齡的距離差為
x1['age']-x2['age']
```

```python
#而cp胸痛指數的距離差為
x1['cp']-x2['cp']
```

兩者的差距為26倍，非常不合理。原因是因為各個欄位的數值尺度不同。

```python
#年齡的最大與最小差距
x['age'].max()-x['age'].min()
```

```python
#cp胸痛指數的最大與最小差距
x['cp'].max()-x['cp'].min()
```

因此在將資料送進預測模型前，需要進行資料的正規化。將資料的數值尺度統一。

## 簡易的方法
最簡單的正規化方法是找出欄位的最大與最小值，然後將最小值當作0而最大值為1，將其他值等距分佈到0~1之間

```python
(x - x.min()) / (x.max()-x.min())
```

這個方法的優點是簡單與直覺，但是容易受到離群值影響。

```python
sns.displot(data,x='thalach')
```

譬如 thalach 這個欄位有很少數的值會小於在90以後

## Z值
使用統計中的標準差(STD)與平均值(MEAN)來調整分數

Z = (X - mean) / std

如此就可以保證大部分的值落於 -3~3 之間，同時也不會犧牲離群值的特異性。

```python
xz = (x - x.mean()) / x.std()
xz
```

```python
sns.displot(xz,x='thalach')
```

## 四分距
有些演算法會需要將值正規化到0~1之間，則Z分數會不太合適。

```python
sns.boxplot(data=data,x='thalach')
```

```python
Q1 = np.percentile(x['thalach'], 25)
Q3 = np.percentile(x['thalach'], 75)
QD = Q3-Q1
QMAX = Q3 + QD*1.5
QMIN = Q1 - QD*1.5
print('Q1=',Q1)
print('Q3=',Q3)
print('QD=',QD)
print('QMAX=',QMAX)
print('QMIN=',QMIN)
```

- 將 QMAX 當作1，QMIN當作0，其他值等距正規化到0~1之間。
- 視需求可將離群值則強制為0或1。

```python
Q1 = x.describe().loc['25%']
Q3 = x.describe().loc['75%']
QD = Q3-Q1
xq = (x - (Q1-1.5*QD))/(3*QD)
xq
```

fps這個欄位的值本來就是 0 或 1，所以不可以也不需要使用這種正規化方法。
