## 神經元的學習過程

```python
# 這裡先產生一份測試用的資料
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = 2 * np.random.rand(100)
y = 4 + 3 * x + np.random.randn(100)

# 一開始將 w 跟 b隨機設定
w = 2
b = 2
u = 0.1 #學習率
```

```python
# 先計算第一筆資料預測值
o = x[0]*w + b
print(f'{o} = {x[0]} * {w} + {b}')
```

```python
# 計算第一筆資料的預測值跟實際值的差異
diff = y[0] - o
print(f'{diff} = {y[0]} - {o}')
```

```python
# 根據誤差調整 w 跟 b 值
# diff = 差越大，w調整越大，預測完全正確時 W = 0 

w_new = w + w*diff * u
b_new = b + 1*diff * u
print(f'{w_new} = {w} * {diff} * {u}')
print(f'{b_new} = 1 * {diff} * {u}')
```

```python
# 用所有的資料去調整權重
w = 2
b = 2
u = 0.01 #學習率很重要，會影響學習速度以及是否產生震盪
for i in range(0,len(x)):
    o = x[i]*w + b
    diff = y[i] - o
    w += w*diff * u
    b += 1*diff * u
    print(f'd = {diff}, w = {w}, b = {b}')
```

```python
# 多次反覆的測試所有資料並調整權重
w = 2
b = 2
u = 0.01
for n in range(0, 20):
    for i in range(0,len(x)):
        o = x[i]*w + b
        diff = y[i] - o
        w += w*diff * u
        b += 1*diff * u
    print(f'w = {w}, b = {b}')
```
