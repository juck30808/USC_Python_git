# 統計常用的觀測指標

```python
# 準備測試用資料
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
bmi_health = pd.read_csv("data/500_Person_Gender_Height_Weight_Index.csv")
height = bmi_health[bmi_health["Index"] == 3]['Height']
weight = bmi_health[bmi_health["Index"] == 3]['Weight']
```

```python
height
```

```python
weight
```

## 平均值

```python
np.average(weight)
```

# 四分位距

```python
# 計算四分位距的方式
q75, q50, q25 = np.percentile(weight, [75 ,50, 25])
print(q75, q50, q25)
```

```python
# 畫出四分位距的方法 Matplotlib
import matplotlib.pyplot as plt
plt.boxplot(weight)
```

```python
# 畫出四分位距的方法 Matplotlib
import matplotlib.pyplot as plt
plt.boxplot(height)
```

```python
# 畫出四分位距的方法 Seaborn
import seaborn as sns
sns.boxplot(weight)
```

## 標準差 Standard Deviation
[標準差 維基百科](https://zh.wikipedia.org/wiki/%E6%A8%99%E6%BA%96%E5%B7%AE)


μ 為平均值, N 為資料總筆數

### 標準差與常態分布


```python
# 計算標準差
np.std(weight)
```

```python
# 產生常態分佈的隨機測試資料，並畫出直方圖
import numpy as np
import matplotlib.pyplot as plt
mean, sigma = 0, 15
x = mean + sigma * np.random.randn(10000)
print('x的標準差為:',np.std(x))
plt.hist(x, bins=50)
```
