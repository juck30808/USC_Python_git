```python
round(data.groupby('Gender')[['MonthlyIncome', #月薪
                              'PercentSalaryHike',  #工資漲幅
                              'Education', 
                              'JobLevel',
                             'TotalWorkingYears']].mean(),2)
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
iris = pd.read_csv('data/iris.csv')
iris

# data = pd.read_csv("data/IBM_HR.csv")
# data.head()
```

```python
iris_data = iris.SepalLengthCm
iris_label = iris.Species
```

```python
train_data , test_data , train_label , test_label = train_test_split(iris_data,
                                                                     iris_label,
                                                                     test_size=0.2)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(train_data,
        train_label)
```

```python
print(knn.predict(test_data))
```
