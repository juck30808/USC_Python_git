```python
import csv
with open('data/test/names.csv', 'w', newline='') as csvfile:
# csvfile = open('data/test/names.csv', 'w', newline='')
    names = ['first_name', 'last_name']
    writer = csv.DictWriter(csvfile, fieldnames=names)

    writer.writeheader()
    writer.writerow({'first_name': 'Baked', 'last_name': 'Beans'})
    writer.writerow({'first_name': 'Lovely', 'last_name': 'Spam'})
    writer.writerow({'first_name': 'Wonderful', 'last_name': 'Spam'})
    print("write!")

print('END')
```

## EXCEL
https://openpyxl.readthedocs.io/en/stable/index.html

```python
from openpyxl import load_workbook
# 預設可讀寫，若有需要可以指定write_only和read_only為True
# wb = load_workbook('data/Iris_short.xlsx', read_only = True)
wb = load_workbook('data/iris.xlsx')

# 獲得所有sheet的名稱
print(wb.sheetnames)
# 根據sheet名字獲得sheet
a_sheet = wb['Iris']
# 獲得sheet名
print(a_sheet.title)
# 獲得當前正在顯示的sheet, 也可以用wb.get_active_sheet()
sheet = wb.active 
print(sheet.title)
```

```python
cell = sheet['B4']
print(cell.column) #B
print(cell.row)    #4
print(cell.value)  #B4 = 4.7

cell = sheet.cell(row=4, column=2)  #B4
print(cell.value)   #4.7
```

```python
# 獲得最大列和最大行
print(sheet.max_row)
print(sheet.max_column)
```

```python
# 因為按行，所以返回A1, B1, C1這樣的順序
for row in sheet.rows:
    for cell in row:
        print(cell.value)

# A1, A2, A3這樣的順序
for column in sheet.columns:
    for cell in column:
        print(cell.value)
```

```python
for cell in list(sheet.rows)[2]:
    print(cell.value)
```

```python
for cell in list(sheet.columns)[2]:
    print(cell.value)
```

```python
for i in range(1, 4):
    for j in range(1, 3):
#         print(i,j)
        print(sheet.cell(row=i, column=j).value)
```

```python
for row_cell in sheet['A1':'B3']:  #逐列但是指定範圍
    for cell in row_cell:
        print(cell.value)
```

```python
from openpyxl import Workbook
wb = Workbook()
sheet = wb.create_sheet("Mysheet")
sheet.cell(row=2, column=2).value = 22
sheet['A1'].value = 11
wb.save('data/test/test.xlsx')
```
