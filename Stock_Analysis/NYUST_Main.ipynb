```python
#3-1  regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

#範例資料
X_actual= np.array([6,8,10,14,18])
y_actual= np.array([7,9,13,17.5,20])

#呼叫回歸函數 將實際帶入做回歸
p1=np.poly1d(np.polyfit(X_actual,y_actual,1))
p2=np.poly1d(np.polyfit(X_actual,y_actual,2))
p3=np.poly1d(np.polyfit(X_actual,y_actual,3))
xp=np.linspace(0,25,100) #0-25產生線性100的值

plt.figure(figsize=(10,5))
plt.plot(X_actual,y_actual,"k.",label=u"data") #黑點位置

plt.plot(xp,p1(xp),'blue' ,label=u"一次線性回歸")
plt.plot(xp,p2(xp),'red',label=u"二次線性回歸")
plt.plot(xp,p3(xp),'black',label=u"三次線性回歸")

#繪圖參數
plt.title("Pizza size/price")
plt.xlabel("size(inch)")
plt.ylabel("price(usd)")
plt.axis([0,25,0,25]) #X軸與y軸尺度
plt.grid(True) #顯示網格
plt.show()
print("一次線性回歸 r-squared:", r2_score(y_actual,p1(X_actual)))
print("二次線性回歸 r-squared:", r2_score(y_actual,p2(X_actual)))
print("三次線性回歸 r-squared:", r2_score(y_actual,p3(X_actual)))
```
