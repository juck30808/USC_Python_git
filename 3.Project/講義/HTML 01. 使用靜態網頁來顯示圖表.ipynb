```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
```

## Matplotlib 圖表存成JPEG

```python
iris = pd.read_csv("Iris.csv")
if not os.path.exists('output'):
    os.mkdir('output')
```

```python
sns.scatterplot(data = iris, x="SepalLengthCm", 
                y="SepalWidthCm", hue="Species")
# 將圖形存成圖檔
# dpi 可以控制圖形大小
plt.savefig('output/image_01.jpg', dpi=300)
```

## 使用靜態網頁連結並顯示圖表
打開 HTML_01.html

## 多張圖表時

```python
plt.close()
sns.kdeplot(data=iris, x="SepalLengthCm", hue="Species")
plt.savefig('output/image_02.jpg', dpi=300)
```

打開 HTML_02.html

## 使用表格來排版圖表

```python
plt.close()
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
plt.savefig('output/image_03.jpg', dpi=300)
```

```python
plt.close()
sns.PairGrid(iris.drop("Id", axis=1), hue="Species"
            ).map(plt.scatter).add_legend()
plt.savefig('output/image_04.jpg', dpi=300)
```
